**Question** **1**: A die is rolled. What is the probability of getting:
(a) An even number
(b) A number greater than 4

**Answer:** Total outcomes in a single die roll = $\{1, 2, 3, 4, 5, 6\}$ (Total = 6)

(a) Even numbers are $\{2, 4, 6\}$.$P(\text{Even}) = \frac{3}{6} = \frac{1}{2} = 0.5$

 (b) Numbers greater than 4 are $\{5, 6\}$.$P(>4) = \frac{2}{6} = \frac{1}{3} \approx 0.333$

 **Question 2:** In a class of 50 students: 20 like Mathematics (M), 15 like Science (S), 5 like both subjects. What is the probability that a student chosen at random likes Mathematics or Science?

**Answer**: We use the Addition Rule for probability: $P(M \cup S) = P(M) + P(S) - P(M \cap S)$$P(M) = \frac{20}{50}$$P(S) = \frac{15}{50}$$P(M \cap S) = \frac{5}{50}$$P(M \cup S) = \frac{20 + 15 - 5}{50} = \frac{30}{50} = \frac{3}{5} = 0.6$


**Question 3**: A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?

**Answer** : Total initial balls = 5 (3 Red, 2 Blue).After drawing one red ball, remaining Red balls = 2.Remaining total balls = 4.The probability that the second ball is red = $\frac{2}{4} = \frac{1}{2} = 0.5$

**Question 4:** The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?

**Answer:**
I would use Stratified Sampling.

Reason: Simple Random Sampling relies on chance and could accidentally result in a sample that doesn't represent both genders equally. By using Stratified Sampling, I can divide the population into two "strata" (Boys and Girls) and then pick an equal number of individuals from each group (e.g., 50 boys and 50 girls) to ensure the 50/50 representation required, regardless of their original 60/40 split in the population.

**Question 5:** The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm. Find the sampling error.

**Answer:**

Sampling Error = $\text{Population Mean} (\mu) - \text{Sample Mean} (\bar{x})$Sampling Error = $160 \text{ cm} - 158 \text{ cm} = 2 \text{ cm}$

**Question 6:** The population mean salary is ₹50,000 with $\sigma$ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?

**Answer:**The formula for Standard Error is $SEM = \frac{\sigma}{\sqrt{n}}$$\sigma = 5000$$n = 100$$SEM = \frac{5000}{\sqrt{100}} = \frac{5000}{10} = ₹500$

**Question 7:** In a group of 100 students: 40 like Cricket (C), 30 like Football (F), 10 like both Cricket and Football. Find the probability that a student likes at least one sport.

**Answer:**"At least one sport" means Cricket OR Football ($C \cup F$).$P(C \cup F) = P(C) + P(F) - P(C \cap F)$$P(C \cup F) = \frac{40}{100} + \frac{30}{100} - \frac{10}{100} = \frac{60}{100} = 0.6$

**Question 8:** From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?  

**Answer:** Probability of first Ace = $\frac{4}{52}$Probability of second Ace = $\frac{3}{51}$ (since one Ace and one card are now gone)$P(\text{Both Aces}) = \frac{4}{52} \times \frac{3}{51} = \frac{1}{13} \times \frac{1}{17} = \frac{1}{221} \approx 0.0045$

**Question 9:** A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?

**Answer:**  Probability of a bulb being defective ($p$) = 0.02Probability of a bulb being non-defective ($q$) = $1 - 0.02 = 0.98$Since the choices are independent:$P(\text{All 5 non-defective}) = (0.98)^5 \approx 0.9039$ or 90.4%

**Question 10:** Differentiate between discrete and continuous random variables with examples.

**Answer:**

A Discrete Random Variable is a variable that has a countable number of distinct values. These usually result from counting and cannot have values between the distinct points (like decimals).

Nature: Obtained by counting.

Values: Whole numbers/integers.

Example 1: The number of students present in a class (you can have 25 or 26 students, but not 25.5).

Example 2: The number of heads obtained in 10 coin tosses.


A Continuous Random Variable is a variable that can take an infinite number of possible values within a given range or interval. These usually result from measurements and can include decimals or fractions.

Nature: Obtained by measuring.

Values: Real numbers (decimals).

Example 1: The height of a student (e.g., 160.5 cm or 172.33 cm).

Example 2: The time it takes to complete a data analytics assignment.
